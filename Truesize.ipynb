```python
import geopandas as gpd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

# Load the world shapefile from your local path
world = gpd.read_file(r"G:\PycharmProjects\numbers\ne_110m_admin_0_countries\ne_110m_admin_0_countries.shp")

# Set up the plot with a Mollweide projection
fig, ax = plt.subplots(figsize=(10, 6), subplot_kw={'projection': ccrs.Mollweide()})

# Plot the world map with Cartopy's Mollweide projection
ax.set_global()  # Show the entire globe
ax.coastlines()  # Add coastlines for context

# Plot the world data using GeoPandas
world.plot(ax=ax, transform=ccrs.PlateCarree(), color='lightgreen', edgecolor='black')

# Save the figure
plt.savefig("world_map1.png", bbox_inches='tight', dpi=300)  # Change the filename and format as needed

# Add title and show the plot
plt.title("World Map in True Size (Mollweide Projection)")
plt.show()

```
