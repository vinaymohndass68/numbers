```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the shapefile for world countries (replace 'path_to_shapefile' with your actual path)
world = gpd.read_file(r"G:\PycharmProjects\numbers\ne_110m_admin_0_countries\ne_110m_admin_0_countries.shp")

# Plot the world map (outlines of countries)
world.boundary.plot(figsize=(10, 6), linewidth=1, edgecolor='black')

# Add a title and show the plot
plt.title("Outlines of World Countries", fontsize=15)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
print(world.columns)
```

```python
# Filter out India using the 'SOVEREIGNT' column
india = world[world['SOVEREIGNT'] == "India"]

# Plot the boundary of India
india.boundary.plot(figsize=(8, 8), linewidth=2, edgecolor='blue')

plt.title("Outline of India")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

# Load the world shapefile (replace with your actual path)
world = gpd.read_file(r"G:\PycharmProjects\numbers\ne_110m_admin_0_countries\ne_110m_admin_0_countries.shp")

# Set up the plot with a Mollweide projection
fig, ax = plt.subplots(figsize=(10, 6), subplot_kw={'projection': ccrs.Mollweide()})

# Plot the world map (outlines of countries)
world.boundary.plot(ax=ax, linewidth=1, edgecolor='black')

# Add titles and labels
plt.title("World Map")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

# Save the figure
plt.savefig("world_map.png", bbox_inches='tight', dpi=300)  # Change the filename and format as needed

# Show the plot
plt.show()
```
